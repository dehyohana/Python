```python
##correlação próx a -1 ou 1, as variáveis são fortemente relacionadas; mais próx a 0 indica pouca correlação
##criar mapa de correlação
import numpy as np

# Compute the correlation matrix
corr = dados.loc[:,'g_0':'g_50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
#Aula 5
#Dado os resultados do experimento, podemos prever se há ao menos um método de ativação?
##scklearn
# f(x) = Ax; x: tipo genético e tipo celular (assinatura celular); y: resposta (ativo/não ativo)
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_MOA']
#treinar o modelo
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2,stratify= y, random_state=7657687)
modelo_rlogistic = LogisticRegression(max_iter=1000)
modelo_rlogistic.fit(x_treino,y_treino)
#receber uma amostra para verificar viabilidade da classificação do modelo através de estabelicimento de métricas
modelo_rlogistic.score(x_teste, y_teste)
```

```python
#implementar modelos mais simples
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score
modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
```

```python
#modelo não diferiu tanto do modelo mais simples de maior frequencia
dados_combinados['ativo_MOA'].value_counts(normalize=True)
```

```python
##Modelo de arvore de decisão
from sklearn.tree import DecisionTreeClassifier
x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_MOA']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2,stratify= y, random_state=7657687)
modelo_arvore = DecisionTreeClassifier(max_depth = 3)
modelo_arvore.fit(x_treino, y_treino)
modelo_arvore.score(x_teste, y_teste)

```

```python
from sklearn import tree
fig,ax = plt.subplots(figsize=(15,10), facecolor = 'k')
tree.plot_tree(modelo_arvore,
               ax=ax,
               fontsize=10,
               rounded=True,
               filled=True,
               feature_names=x_treino.columns,
               class_names=['Não ativado','Ativado'])
plt.show()
```

```python
##Modelo de arvore de decisão
from sklearn.tree import DecisionTreeClassifier
x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_MOA']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2,stratify= y, random_state=376)

teste = []
treino = []
#adicionando profundidade e verificando o quão bem está acertando
for i in range(1,15):
  modelo_arvore = DecisionTreeClassifier(max_depth=i)
  modelo_arvore.fit(x_treino, y_treino)
  teste.append(modelo_arvore.score(x_teste, y_teste))
  treino.append(modelo_arvore.score(x_treino, y_treino))
```
